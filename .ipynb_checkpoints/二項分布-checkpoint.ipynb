```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

#離散二項分布(実験)
# 確率pで成功する試行をn回行った時に、x回(x軸)成功する確率(y軸)の分布
def Binomial_EX(n,p):
    sample = np.random.rand(n)
    return len(np.where(sample < p)[0])

fig,axes = plt.subplots(ncols=3,nrows=2,figsize=(15,10))
for p,i in zip([0.3,0.5,0.9],range(3)):
    result = {i:0 for i in range(101)}
    for j in range(1000):
        result[Binomial_EX(100,p)] += 1
    sumval = sum(result.values())
    x = list(range(101))
    y = list(map(lambda x:x/sumval,result.values()))
    title = 'B(100,{})'.format(p)
    axes[0,i].bar(x,result.values(),width=0.5)
    axes[0,i].set_title(title)
    axes[0,i].set_xlabel('X')
    axes[0,i].set_ylabel('freq')
    axes[1,i].bar(x,y,width=0.5)
    axes[1,i].set_title(title)
    axes[1,i].set_xlabel('X')
    axes[1,i].set_ylabel('probability')
    
    
```

```python
#離散二項分布(理論)
from scipy.misc import comb
f = lambda n,x,p:comb(n,x) * (p ** x) * ((1 - p) ** (n-x))
N = [100] * (101)
x = list(range(101))
fig,axes = plt.subplots(ncols=3,figsize=(15,5))
for p,i in zip([0.3,0.5,0.9],range(3)):
    P = [p] * (101)
    y = list(map(f,N,x,P))
    title = 'B(100,{})'.format(p)
    axes[i].bar(x,y,width=0.5)
    axes[i].set_title(title)
    axes[i].set_xlabel('X')
    axes[i].set_ylabel('probability')
```
