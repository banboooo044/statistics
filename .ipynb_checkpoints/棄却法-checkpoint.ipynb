```python
#2/20/2018
#棄却法のイメージ
#横軸にランダムに位置を決めるY(U[0,1])-> 縦軸にランダムに位置を決める(U[0,M*g(Y)]) -> f(Y)より小さければ採用
#f(Y)の大きさと、採用される数が分布として一致する。 -> Yに対する採用された数が目標の分布fと一致。
```

```python
# ベータ乱数を受理・棄却法で生成
# 目標分布（ここではベータ分布）のpdfは既知とする
# 提案分布として一様分布を使用

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
from scipy.stats import uniform, beta

np.random.seed()

# 目標分布f
f = beta(a=2.7, b=6.3).pdf

# 提案分布g1(一様分布)
gv1 = uniform
g1 = gv1.pdf

#提案分布g2(ベータ分布(a,bが違う))
gv2 = beta(a=2.0, b=6.0)
g2 = gv2.pdf

# 分布の上限を指定する定数Mを設定
# ベータ分布のpdfの上限値を指定すればベータ分布をすべて覆える
# 最大値を求めるためにベータ分布のpdfにマイナスをつけて
# 最小値問題に帰着させる
xopt = scipy.optimize.fmin(lambda x: -f(x), 0.0, disp=False)
M = f(xopt)[0]
print ("M =" + str(M))# M := 確率関数fの最大値

xopt = scipy.optimize.fmin(lambda x: -(f(x)/g2(x)),0.0,disp=False)
C = (f(xopt) / g2(xopt))[0]
print("C =" + str(C))

# 受理・棄却法
Nsim = 10000

# 提案分布gからの乱数Yを生成(横軸の位置をランダムに決定)
Y1 = gv1.rvs(size=Nsim)
Y2 = gv2.rvs(size=Nsim)

# 一様乱数UをNsim個生成(縦軸の位置をランダムに決定)
U1 = M * g1(Y1) * uniform.rvs(size=Nsim) 
U2 = C * g2(Y2) * uniform.rvs(size=Nsim)

acceptedIdx1 = U1 <= f(Y1) 
rejectedIdx1 = U1 > f(Y1)
acceptedIdx2 = U2 <= f(Y2) 
rejectedIdx2 = U2 > f(Y2)

# Yから受理の条件を満たすサンプルXを残して残りを棄却
X1 = Y1[U1 <= f(Y1)]
X2 = Y2[U2 <= f(Y2)]

print("g = U[0,1]の場合")
print ("サンプル数: {0} => {1}".format(len(Y), len(X1)))
print ("受理率 : {0}".format(len(X1) / float(len(Y))))

print("g = {0}*fの場合".format(C))
print ("サンプル数: {0} => {1}".format(len(Y), len(X2)))
print ("受理率 : {0}".format(len(X2) / float(len(Y))))

print("受理率が2つ目の方が高くて、効率的!!!!")

# 目標分布を描画
x = np.linspace(0.0, 1.0, 1000)
y = f(x)
z1 = M * g1(x)
z2 = C * g2(x)

plt.figure(figsize=(60, 30))
plt.subplot(221)
plt.plot(x, y, 'r-', lw=2) # f
plt.plot(x, z1, 'g-', lw=2) # g
plt.title('Beta')
plt.xlabel('X')
plt.ylabel('prob')
plt.hist(X1, bins=50, normed=True)

plt.subplot(222)
plt.plot(x, y, 'r-', lw=2) # f
plt.plot(x, z1, 'g-', lw=2) # g
plt.scatter(Y1[acceptedIdx1], U1[acceptedIdx1], color="red")
plt.scatter(Y1[rejectedIdx1], U1[rejectedIdx1], color="blue")

plt.subplot(223)
plt.plot(x, y, 'r-', lw=2) # f
plt.plot(x, z2, 'g-', lw=2) # g
plt.hist(X2, bins=50, normed=True)

plt.subplot(224)
plt.plot(x, y, 'r-', lw=2) # f
plt.plot(x, z2, 'g-', lw=2) # g
plt.scatter(Y2[acceptedIdx2], U2[acceptedIdx2], color="red")
plt.scatter(Y2[rejectedIdx2], U2[rejectedIdx2], color="blue")

plt.show()
```
