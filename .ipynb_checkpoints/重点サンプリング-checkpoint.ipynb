```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
from scipy.stats import norm,uniform
%matplotlib inline

#  5.0  x 
a = 5.0
ix = np.arange(-5, 15, 0.01)
h = lambda x: x > a
# 
f = norm.pdf
y = lambda x: h(x) * f(x)
plt.figure(figsize=(10,25))
plt.subplot(421)
plt.plot(ix,f(ix),label='f(x)')
plt.title('N(0,1)-integrand[5.0:]')

plt.subplot(422)
plt.plot(ix,f(ix),label='f(x)')
plt.title('<- enlarged')
plt.xlim(4.0,6.0)
plt.ylim(0.0,0.0001)

# scipy.integrate
I1 = scipy.integrate.quad(f, a, np.inf)[0]
I2 = scipy.integrate.quad(y, -np.inf, np.inf)[0]
print ("scipy.integrate1:" + str(I1))
print ("scipy.integrate1:" + str(I2))

N = 10000
# 
# xN(x|0,1)
# 5
# h(x)0
x = norm.rvs(size=N)
I3 = np.mean(h(x))
plt.subplot(423)
plt.title('normal monte carlo')
plt.plot(ix,h(ix),label='f(x)')
plt.plot(ix,f(ix),label='p(x)')
plt.legend()

print("normal monte carlo integration(g(x)=N(5,1)):" + str(I3))

# 
# g(x)5N(x|5,1)
g = norm(loc=5, scale=1).pdf
x = norm(loc=5, scale=1).rvs(size=N)
I4 = np.mean(f(x) / g(x) * h(x))
print ("importance sampling:"+ str(I4))

plt.subplot(425)
plt.plot(ix, g(ix), label="g(x) = p(x)")
plt.plot(ix, (h(ix)*f(ix))/g(ix), label="f*h/g(x)")
plt.legend()

plt.subplot(426)
plt.title('<- enlarged')
plt.plot(ix, g(ix), label="g(x) = p(x) = N(5,1)")
plt.plot(ix, (h(ix)*f(ix))/g(ix), label="f*h/g(x)")
plt.legend()
plt.xlim(4.0,6.0)
plt.ylim(0.0,0.00001)

def convergence(f,p = uniform()):
    N = 1000
    X = f(p.rvs(size=N))
    estint = np.cumsum(X) / np.arange(1,N+1) #((1 ~ N))
    esterr = np.sqrt(np.cumsum((X-estint)**2)) / np.arange(1, N + 1)    
    return estint,esterr

plt.subplot(427)
estint,esterr = convergence(h,norm)
plt.title('normal monte carlo convergence')
plt.plot([I1]*len(estint),color='cyan',label='trueVal')
plt.plot(estint,color='red',label='predict')
plt.plot(estint + 2 * esterr, color='orange')
plt.plot(estint - 2 * esterr, color='orange')
plt.legend()

plt.subplot(428)
func = lambda x:f(x)*h(x)/g(x)
estint,esterr = convergence(func,norm(loc=5, scale=1))
plt.title('importance sampling convergence1')
plt.plot([I1]*len(estint),color='cyan',label='trueVal')
plt.plot(estint,color='red',label='predict')
plt.plot(estint + 2 * esterr, color='orange')
plt.plot(estint - 2 * esterr, color='orange')
plt.legend()

```
