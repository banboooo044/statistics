```python
#2/20/2018
#Box-Muller法 - 標準正規分布を二つの独立な一様分布から作成。
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
%matplotlib inline

# 一様分布に従う確率分布からランダムサンプリング
np.random.seed()
rv = uniform(loc=0.0, scale=1.0)
N = 10000
U1 = rv.rvs(size=N) 
U2 = rv.rvs(size=N)

p = lambda x:(np.exp(-(x**2)/2))/np.sqrt(2*np.pi)
f = lambda U1,U2:np.sqrt((-2)*np.log(U1)) * np.cos(2*np.pi*U2) #変換関数

plt.figure(figsize=(15, 5))
X1 = list(map(f,U1,U2)) # N(0,1)を作成1
X2 = list(map(f,U2,U1)) # N(0,1)を作成2
x = np.linspace(-10, 10, 200)
xmin = -10
xmax = 10
nbins = 50

#1つ目の正規分布
plt.subplot(121)
plt.hist(X1,nbins,normed=True)
plt.title('N(0,1)')
plt.xlabel('X')
plt.ylabel('num')
plt.xlim(xmin, xmax)
plt.plot(x, list(map(p,x)), 'r-', lw=1, label='normal pdf')

#二つ目の正規分布
plt.subplot(122)
plt.hist(X2,nbins,normed=True)
plt.title('N(0,1)')
plt.xlabel('X')
plt.ylabel('num')
plt.xlim(xmin, xmax)
plt.plot(x, list(map(p,x)), 'r-', lw=1, label='normal pdf')

plt.tight_layout()  
plt.show()
```
