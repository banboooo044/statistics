```python
#2/19/2018
#一様乱数からポアソン分布に従う乱数生成

import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
from scipy.misc import factorial
%matplotlib inline

# ポアソン分布の累積確率関数を求める。
lam = 10
xmax = 40
f = lambda lam,x:np.exp(-lam) * (lam ** x) / factorial(x)
acc = list(map(f,[lam]*(xmax),list(range(xmax))))
for i in range(xmax-1):
    acc[i+1] += acc[i]
# acc[i] := p(x)(x<=i)の累積。

# [0.0, 1.0]の一様分布に従う確率変数
np.random.seed()
N = 10000
rv = uniform(loc=0.0, scale=1.0)
sample = np.array(rv.rvs(size=N)) # N(10000)個のサンプリング。

X_acc = [0] * 40
X = []
for i in range(1,xmax):
    X_acc[i] = len(np.where(sample < acc[i])[0])
    X.extend([i] * (X_acc[i] - X_acc[i-1])) # acc[i-1] < sample < acc[i] となる時、p_inv(sample) = i

nbins = 50
weights = np.ones_like(X) / float(len(X))
plt.hist(X,nbins, weights=weights)
plt.plot(list(range(40)),list(map(f,[lam]*40,list(range(40)))),'ro-', lw=1)
plt.xlabel('X')
plt.ylabel('probabillity')
plt.title('Po(10)')
plt.show()
```
